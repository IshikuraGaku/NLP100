```python
import re
import json
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
f = open("neko_dic.json", "rt", encoding="utf-8")
```

```python
analitics_list = json.load(f)
print(type(analitics_list.values()))
```

```python
analitics_list = list(analitics_list.values())
```

```python
 count_dic = {}
for dic in analitics_list:
    if count_dic.get(dic.get("base")) is not None:
        count_dic[dic.get("base")] = count_dic[dic.get("base")] + 1
    else:
        count_dic[dic.get("base")] = 1
```

```python
count_dic = sorted(count_dic.items(), key=lambda x:x[1], reverse=True)
```

```python
count_dic
```

```python
top10 = count_dic[:10]
```

```python
top10_word = []
top10_num = []
for w, n in top10:
    top10_word.append(w)
    top10_num.append(n)
print(top10_word)
```

```python
plt.xlabel("    ".join(top10_word), fontname="MS Gothic")
plt.bar(np.array(range(10)), np.array(top10_num))
```
